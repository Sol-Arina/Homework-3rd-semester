#### Задача 1 (10 баллов). 

Представим себе, что мы хотим открыть музыкальный магазин и заведовать всеми его делами. Напишем программу, которая будет визуализировать наши мечты. Наверное, в первую очередь нам понадобится класс для самого магазина. У магазина есть счет, склад, возможность продавать инструменты и покупать их (может быть, что-то еще - что придумаете). Потом, нужен, конечно, класс для музыкального инструмента. Попробуйте реализовать музыкальные инструменты с помощью наследования: можно создать некий абстрактный класс "инструмент", а от него будут наследовать классы "гитара", "пианино" и так далее - при этом у гитар и пианино могут быть свои разные характеристики, н-р, у гитары количество струн, тип порожка и так далее. Можно и более подробную иерархию, если будете успевать (например, инструменты бывают электрическими и акустическими). Наконец, можно создать класс "покупатель", у которого будет кошелек. 

Наконец, неплохо бы не забыть написать некий код, который будет приводить наши классы в движение: имитировать рабочий день магазина (можно в цикле), с рандомной вероятностью посещение покупателей, давать нашему владельцу-пользователю возможность решить, не пора ли закупить новые инструменты или, может быть, затеять рекламную кампанию. 

Фантазия всячески поощряется, могу за нее поставить дополнительные баллы. 

```python
import time
from random import randint
import random
```

```python
# класс для муз инструментов
class Instrument:

    def __init__(self, brand, price):
        self.brand = brand
        self.price = price
    

# гитара
class Guitar(Instrument):
    tag = 'Гитара'
    def __init__(self, brand, price, string_count, acou_or_ele):
        self.brand = brand
        self.price = price
        self.string_count = string_count
        #self.fret_type = fret_type
        self.acou_or_ele = acou_or_ele

# пианино
class Piano(Instrument):
    tag = 'Пианино'
    def __init__(self, brand, price, key_count, acou_or_ele):
        self.brand = brand
        self.price = price
        self.key_count = key_count
        self.acou_or_ele = acou_or_ele 

# скрипка
class Violin(Instrument): 
    tag = 'Скрипка'
    def __init__(self, brand, price, size, acou_or_ele):
        self.brand = brand
        self.price = price
        self.size = size #Apart from the standard full (4⁄4) size, violins are also made in so-called fractional sizes of 7⁄8, 3⁄4, 1⁄2, 1⁄4, 1⁄8, 1⁄10, 1⁄16,
        self.acou_or_ele = acou_or_ele # Classical Violin vs Electric Violin


# гуцинь
class Guqin(Instrument):
     tag = 'Гуцинь'
     def __init__(self, brand, price, string_type):
        self.brand = brand
        self.price = price
        self.string_type = string_type  # silk (if it's old) OR nylon-flatwound steel strings



# ксилофон
class Xylophone(Instrument):
     tag = 'Ксилофон'
     def __init__(self, brand, price, key_count, type):
        self.brand = brand
        self.price = price
        self.key_count = key_count
        self.type = type
        self.variety = ['diatonic', 'chromatic', 'bass', 'concert']


# покупатель
class Customer:

    def __init__(self, name, wallet):
        self.name = name
        self.wallet = wallet
        self.happiness = 0    

    def buy_instrument(self, instrument, shop):
        purchase_prob = 0.5  # базовая вероятность покупки 50 %
        if random.random() > 0.4:  # 60 % что покупатель зайдёт в магаз
            print(f'покупатель {self.name} решает зайти в магазин {shop.name}')
            if random.random() > purchase_prob:
                print(f'Покупатель {self.name} смотрит разные инструменты...') 
                if self.wallet >= instrument.price:
                    if instrument in shop.warehouse:
                        self.wallet -= instrument.price
                        print(f'{self.name} решает приобрести {instrument.tag} марки {instrument.brand} за {instrument.price} монет.')
                        shop.sell_instruments(instrument)
                        print('успешная продажа')
                    else:
                        print(f'У {self.name} не получилось купить {instrument.tag}, его нет в наличии в магазине.')
                else:
                    print(f'У {self.name} недостаточно средств, чтобы купить {instrument.tag}.')


# тайный покупатель
class MysteryShopper:
    def __init__(self, name):
        self.name = name        
        self.shop_score = 0
        

    def give_scoring(self, shop, customer):
        if len(shop.warehouse) >= 1 and shop.cleanliness == True and shop.workers >= 1: # проверка на пустой лист
            self.shop_score = randint(7,10)
            customer.purchase_prob = 0.8
        elif len(shop.warehouse) >= 1 and shop.cleanliness == False and shop.workers >= 1:
            self.shop_score = randint(4,6)
            customer.purchase_prob = 0.6
        else:
            self.shop_score = randint(0,3)
            customer.purchase_prob = 0.3


    def perhaps(self, shop, customer):
        if random.random() < 0.2:  # 20% вероятность проверки
            print(f'Тайный покупатель {self.name} зашел в магазин, какую же оценку он поставит?')
            self.give_scoring(shop, customer)
            print(f'Тайный покупатель оценил ваш магазин на {self.shop_score} баллов')
        else:
            print(f'Тайный покупатель {self.name} не зашел в этот раз... УРА!!')



# МАГАЗИН
class MusicShop:
    def __init__(self, name):
        self.name = name
        self.bank_account = 0
        self.warehouse = []
        self.cleanliness = False
        self.workers = 0 
      

    def stocking_warehouse(self, instrument):
        '''закупки на склад'''
        if self.bank_account >= instrument.price:
            self.warehouse.append(instrument)
            self.bank_account -= instrument.price
            print(f'На склад поступил товар - {instrument.tag} марки {instrument.brand} за {instrument.price} монет.')
        else:
            print(f'У магазина {self.name} недостаточно средств для покупки этого товара :(') 


    def sell_instruments(self, instrument):
        '''for selling instruments'''
        if instrument in self.warehouse:
            self.warehouse.remove(instrument)
            self.bank_account += instrument.price
            print(f'Магазин продал {instrument.tag} за {instrument.price}. Текущий баланс магазина: {self.bank_account}')
        else:
            print(f'{instrument.tag} отсутствует в наличии для продажи.')
        
    def look_through_warehouse(self):
        '''посмотреть, что есть на складе'''
        if self.warehouse:
            print('Сейчас в наличии:')
            for instrument in self.warehouse:
                print (f'{instrument.tag} марки {instrument.brand}, цена - {instrument.price} монет')
        else:
            print('На складе нет инструментов...')            


# работник ОБЩИЙ
class Employee:
    def __init__(self, name, age, has_kids):
        self.name = name
        self.age = age
        self.has_kids = has_kids # если есть дети, то вероятность пропустить работу чуть больше
        self.happiness = 0
        self.missed_work = True

    def come_to_work(self, shop):
        if self.has_kids == False and random.random() > 0.05: # 95 % что работник придёт на работу 
                self.missed_work = False
                shop.workers += 1
        elif self.has_kids == True and random.random() > 0.15 : # 85 % что работник придёт на работу 
                self.missed_work = False
                shop.workers += 1


# уборщик
class Cleaning(Employee):
    def clean_shop(self, shop):
        print(f'Уборщик {self.name} сделал уборку в торговом зале!')
        shop.cleanliness = True
        

# продавец-консультант
class Salesperson(Employee):
    def chat_with_customer(self, customer):
        print(f'Продавец-консультант {self.name} поговорил с покупателями!')
        customer.purchase_prob = 0.7
    

# спец по рекламе, которого нанимают кое-когда
class Promotion(Employee):
    def promotion_campaign(self, customer):
        print('Специалист по рекламе запустил рекламную кампанию!')
        customer.purchase_prob = 0.9
```

```python
def workday_sim(shop, customers, salespeople, cleaner, unusual_customer, customer): 
    print(f'Начинается новый рабочий день. Добро пожаловать в {shop.name}')

    shop.look_through_warehouse()

    # customers are spawning, business is booming
    for customer in customers:
        instrument_for_purchase = random.choice(shop.warehouse) if shop.warehouse else None
        if instrument_for_purchase:
            customer.buy_instrument(instrument_for_purchase, shop)
            if instrument_for_purchase not in shop.warehouse:
                shop.sell_instruments(instrument_for_purchase)
            else:
                print('Покупатель ушёл без покупки')    

    # уборщик пришел
    cleaner.come_to_work(shop)
    cleaner.clean_shop(shop)   
    
    # продавцы пришли
    for salesperson in salespeople:
        salesperson.come_to_work(shop)
        salesperson.chat_with_customer(customer)
      
    unusual_customer.perhaps(shop, customer)  # заходит тайник... может заходит, а может и нет 
                
    # конец дня заполнение склада
    decide = input(f'Хотите закупить новые инструменты для магазина {shop.name} (да/нет)?') # да нет
    if decide.lower() == 'да':
        chosen_instruments = input('Какой инструмент вы хотите купить? (гитара/пианино/скрипка/гуцинь/ксилофон)')
        if chosen_instruments.lower() == 'гитара':
            brand = input('Введите бренд:')
            price = int(input('Введите цену:'))
            string_count = int(input('Введите кол-во струн: (от 6 до 18)'))
            acou_or_ele = input('Акустика или электро-? (акустика/электро):')
            new_guitar = Guitar(brand, price, string_count, acou_or_ele)
            shop.stocking_warehouse(new_guitar)   

        elif chosen_instruments.lower() == 'пианино':
            brand = input('Введите бренд:')
            price = int(input('Введите цену:'))
            key_count = int(input('Введите кол-во клавиш: '))
            acou_or_ele = input('Акустика или электро-? (акустика/электро):')
            new_piano = Piano(brand, price, key_count, acou_or_ele)
            shop.stocking_warehouse(new_piano)       

        elif chosen_instruments.lower() == 'скрипка':
            brand = input('Введите бренд:')
            price = int(input('Введите цену:'))
            size = float(input('Введите размер (0.8/0.75/0.25/0.5): ')) # self.size = size #Apart from the standard full (4⁄4) size, violins are also made in so-called fractional sizes of 7⁄8, 3⁄4, 1⁄2, 1⁄4, 1⁄8,
            acou_or_ele = input('Акустика или электро-? (акустика/электро):') #self.acou_or_ele = acou_or_ele # Classical Violin vs Electric Violin  
            new_violin = Violin(brand, price, size, acou_or_ele)
            shop.stocking_warehouse(new_violin) 

        elif chosen_instruments.lower() == 'гуцинь':
            brand = input('Введите бренд:')
            price = int(input('Введите цену:'))
            strings_type = input('Введите тип струн (шёлк/нейлон): ') # silk (if it's old) OR nylon-flatwound steel strings
            new_qin = Guqin(brand, price, strings_type)
            shop.stocking_warehouse(new_qin)

        elif chosen_instruments.lower() == 'ксилофон':
            brand = input('Введите бренд:')
            price = int(input('Введите цену:'))
            new_xyl = Xylophone(brand, price, type)
            new_xyl.type = random.choice(new_xyl.variety)
            shop.stocking_warehouse(new_xyl)

    # нанять специалиста по рекламе
    promo_decision = input(f'Хотите заказать рекламу для магазина {shop.name}? (да/нет)')
    if promo_decision.lower() == 'да':
        promo_employee = Promotion('Melissa', 31, False)
        promo_employee.promotion_campaign(customer)
    else:
        print('Вы решили обойтись без рекламы')                                                                   
            
    print('Рабочий день закончился. Магазин закрывается.')        
        


```

```python
# START NEW GAME

if __name__ == '__main__':

    shop = MusicShop('Do you like Jazz?')
    shop.bank_account = 15000

    # покупатели 
    customers = [Customer('Jinny', 675), 
                 Customer('Alex Vorobiev', 1100), 
                 Customer('Crystal', 1312), 
                 Customer('The Smashing Pumpkins', 1700),
                 Customer('Unknown Local Experimental Indie Synth Rock Band', 100),
                 Customer('Eddie', 1467),
                 Customer('Lady Gaga', 3000),
                 Customer('Alice from the local orchestra', 2400),
                 Customer('Brad the music journalist', 1000)]
    
    customer = Customer('Petya', 500)
    
    unusual_customer = MysteryShopper('Larisa Dolina')
    
    # работяги
    salespeople = [Salesperson('Danny', 22, False),
                   Salesperson('Nicole', 37, True)]
    cleaner = Cleaning('Vasya', 29, True)

```

```python
for day in range(1, 4):
    print(f'\n*** День {day} ***')
    workday_sim(shop, customers, salespeople, cleaner, unusual_customer, customer)
    time.sleep(3)
    print('-- Спокойной ночи --')
print(f'\nБаланс счёта магазина после этой рабочей недели - {shop.bank_account} монет')    
```
